## AISHELL-1

```python
path_bart = '/home/users/jiangjin/jiangjin_bupt/ASR_CORRECTION/Context_Correction/context_seq2seq/T-model-baseline/AISHELL-1.new.v2.final/result/T_modal_AISHELL-1.txt'
path_AEC = '/home/users/jiangjin/jiangjin_bupt/ASR_CORRECTION/Context_Correction/context_seq2seq/T-model-knn-ckpt/zh-AISHELL-1.sos.db.v1.final/result/T_modal_AISHELL-1.txt'
```

```python
with open(path_bart, 'r') as f_bart:
    data_bart = f_bart.readlines()
    data_bart_record = [item.strip().split(' ')[0] for item in data_bart]
    data_bart_label = [item.strip().split(' ')[1] for item in data_bart]
    
```

```python
with open(path_AEC, 'r') as f_AEC:
    data_AEC = f_AEC.readlines()
    data_AEC_record = [item.strip().split(' ')[0].replace('$','') for item in data_AEC]
    data_AEC_label = [item.strip().split(' ')[1].replace('$','') for item in data_AEC]
    
```

```python
len(data_bart_label), len(data_AEC_label)
```

```python
data_AEC_record_new = []
data_AEC_label_new = []
for item in data_bart_label:
    index = data_AEC_label.index(item)
    data_AEC_record_new.append(data_AEC_record[index])
    data_AEC_label_new.append(data_AEC_label[index])
```

```python
data_AEC_label_new[10],data_bart_label[10]
```

```python
from datasets import load_metric
metric = load_metric('cer')
```

```python
data_bart_cer = [metric.compute(predictions=[item_x], references=[item_y]) for item_x, item_y in zip(data_bart_record, data_bart_label)]
```

```python
data_AEC_cer = [metric.compute(predictions=[item_x], references=[item_y]) for item_x, item_y in zip(data_AEC_record_new, data_AEC_label_new)]
```

```python
for i in range(len(data_AEC_cer)):
    bart_cer = data_bart_cer[i]
    AEC_cer = data_AEC_cer[i]
    if bart_cer>AEC_cer:
        print(i)
        print('---label----')
        print(data_AEC_label_new[i])
        print('-------')
        print(data_bart_record[i])
        print(data_AEC_record_new[i])
        print('----------------------------')
```
